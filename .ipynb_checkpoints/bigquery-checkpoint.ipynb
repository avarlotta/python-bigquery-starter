### A quick look at "UNNEST" and BigQuery with Pandas



```python
from google.cloud import bigquery
from google.oauth2 import service_account
from constants import PROJECT_ID
import pandas as pd

# Use a service account for authentication onto GCP
credentials = service_account.Credentials.from_service_account_file('secrets/sa_key_01.json')

# The project id to which the service account has access
project_id = PROJECT_ID

# Establish a connection to the client
client = bigquery.Client(credentials= credentials, project=project_id)

# Perform a query
query = client.query(
"""
SELECT event_date, event_timestamp, event_params.value.string_value as page_location 
FROM `bigquery-public-data.ga4_obfuscated_sample_ecommerce.events_20210131`,
UNNEST(event_params) AS event_params
WHERE event_params.key = 'page_location'
""")

#Convert output into a pandas dataframe
df = query.to_dataframe()
df
```

```python
query2 = client.query("""
SELECT event_date, event_params.value.string_value as page_location, COUNT(page_location) as page_count 
FROM `bigquery-public-data.ga4_obfuscated_sample_ecommerce.events_20210131`,
UNNEST(event_params) AS event_params
WHERE event_params.key = 'page_location'
GROUP BY event_date, page_location
HAVING page_count>100"""
)

df2 = query2.to_dataframe()
```

```python
# Describe page count stats
```

```python
# Plot page count
df2.page_count.plot(kind='hist')
```

```python
import seaborn as sns

sns.distplot(df2.page_count, bins=100)
```
